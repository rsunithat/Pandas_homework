## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels =group_name)
age_count = purchase_data.groupby("Age Group").count()
age_count = age_count[["Age"]]
age_count = age_count.rename(columns={"Age": "Purchase Count"})
age_average_price = purchase_data.groupby("Age Group").mean()
age_average_price = age_average_price[["Price"]]
age_average_price = age_average_price.rename(columns={"Price": "Average Purchase Price"})


merge1 = pd.concat([age_average_price, age_count], axis=1)
age_total_purchase = purchase_data.groupby("Age Group").sum()
age_total_purchase = age_total_purchase[["Price"]]
age_total_purchase = age_total_purchase.rename(columns={"Price": "Total Purchase Value"})
merge2 = pd.concat([merge1, age_total_purchase], axis=1)
merge2["Avg Total Purchase"] = merge2["Total Purchase Value"] / merge2["Purchase Count"]
purchase_analysis_age = merge2[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase"]]
purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender_sum=purchase_data.groupby('SN').sum()
top_spender_count=purchase_data.groupby('SN').count()
top_spender_price=purchase_data.groupby('SN').sum()/purchase_data.groupby('SN').count()
top_spender_df = pd.DataFrame({'Purchase Count':top_spender_count['Price'],
                                'Average Purchase Price':top_spender_price['Price'],
                                'Total Purchase Value':top_spender_sum['Price']})
top_spender_df = top_spender_df.sort_values(by="Purchase Count",ascending=False)
top_spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_count= purchase_data.groupby(['Item ID','Item Name']).count()['Price']
popular_price= purchase_data.groupby(['Item ID','Item Name']).mean()['Price']
popular_sum= purchase_data.groupby(['Item ID','Item Name']).sum()['Price']
popular_df =pd.DataFrame({'Purchase Count':popular_count,
                          'Item Price':popular_price,
                          'Total Purchase Value':popular_sum})
popular_df = popular_df.sort_values(by="Purchase Count",ascending=False)
popular_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profit_count= purchase_data.groupby(['Item ID','Item Name']).count()['Price']
most_profit_price= purchase_data.groupby(['Item ID','Item Name']).mean()['Price']
most_profit_sum= purchase_data.groupby(['Item ID','Item Name']).sum()['Price']
most_profit_df =pd.DataFrame({'Purchase Count':most_profit_count,
                          'Item Price':most_profit_price,
                          'Total Purchase Value':most_profit_sum})
most_profit_df = most_profit_df.sort_values(by="Total Purchase Value",ascending=False)
most_profit_df
```
